```python
def haversine_array(lat1, lng1, lat2, lng2):
    '''
    The distance between the two cordinates on the earth
    '''
    lat1, lng1, lat2, lng2 = map(np.radians, (lat1, lng1, lat2, lng2))
    AVG_EARTH_RADIUS = 6371  # in km
    lat = lat2 - lat1
    lng = lng2 - lng1
    d = np.sin(lat * 0.5) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(lng * 0.5) ** 2
    h = 2 * AVG_EARTH_RADIUS * np.arcsin(np.sqrt(d))
    return h

def direction_array(lat1, lng1, lat2, lng2):
    '''
    This function gets the direction array of the the two locations 
    '''
    AVG_EARTH_RADIUS = 6371  # in km
    lng_delta_rad = np.radians(lng2 - lng1)
    lat1, lng1, lat2, lng2 = map(np.radians, (lat1, lng1, lat2, lng2))
    y = np.sin(lng_delta_rad) * np.cos(lat2)
    x = np.cos(lat1) * np.sin(lat2) - np.sin(lat1) * np.cos(lat2) * np.cos(lng_delta_rad)
    return np.degrees(np.arctan2(y, x))

 
def dummy_manhattan_distance(lat1, lng1, lat2, lng2):
    '''
    This function gets manhatan distance from point 1 to point2
    '''
    a = haversine_array(lat1, lng1, lat1, lng2)
    b = haversine_array(lat1, lng1, lat2, lng1)
    return a + b

complete_data['manhattan_dist'] = dummy_manhattan_distance(complete_data['pickup_latitude'].values, 
                                                     complete_data['pickup_longitude'].values, 
                                                     complete_data['dropoff_latitude'].values, 
                                                     complete_data['dropoff_longitude'].values)



complete_data['haversine_distance'] = haversine_array(complete_data['pickup_latitude'].values, 
                                                     complete_data['pickup_longitude'].values, 
                                                     complete_data['dropoff_latitude'].values, 
                                                     complete_data['dropoff_longitude'].values)


complete_data['direction'] = direction_array(complete_data['pickup_latitude'].values, 
                                          complete_data['pickup_longitude'].values, 
                                          complete_data['dropoff_latitude'].values, 
                                          complete_data['dropoff_longitude'].values)
```
